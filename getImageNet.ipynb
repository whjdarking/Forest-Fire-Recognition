```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jun 30 16:28:26 2017

@author: hello
"""
import requests
import urllib.request
import time#导入包

r=requests.get('http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n11639445')
names=r.text
url=names.split('\r\n')
imgname=[]
for i in range(0,len(url)):
    try:
        print ( url[i]+"->"+str(i+1)+'.jpg')
        imgname.append('D:/image/'+str(i+1)+'.jpg')
        web = urllib.request.urlopen(url[i])
        #time.sleep(3)#设置时间间隔为3秒
        data = web.read()
        f = open(imgname[i],"wb")
        f.write(data)
        f.close()
        print ("->downloaded!")
    except urllib.request.HTTPError as reason:
        print(",fail to download, caused by "+reason.reason)
        continue
    except urllib.request.URLError as reason:
        print(",fail to download, caused by URLError")
        continue
    except:
        print(", FAILED for unknow reason!")
        continue
```
